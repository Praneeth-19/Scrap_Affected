Scrap affected, death, survived/ recover, active cases of corona virus division
state wise in India. Collect data by web scraping from minimum 3 different 
website heading this section with which site data belongs and create some 
chart in python to represent collected data

```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

# Define the URLs of the websites you want to scrape
urls = [
    {
        'url': ' https://www.mohfw.gov.in/',
        'source': 'Ministry of Health and Family Welfare, Government of India'
    },
    {
        'url': 'https://www.icmr.gov.in/',
        'source': 'Indian Council of Medical Research (ICMR)'
    },
    {
        'url': 'https://www.who.int/india',
        'source': 'World Health Organization (WHO) India'
    }
]

# Create empty lists to store the scraped data
states = []
affected = []
death = []
recovered = []
active = []

# Iterate over the URLs
for url_info in urls:
    url = url_info['url']
    source = url_info['source']

    # Send a GET request to the URL
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the relevant elements and extract the data
    # Modify this according to the HTML structure of the websites you are scraping
    state_elements = soup.find_all('td', class_='state')
    affected_elements = soup.find_all('td', class_='affected')
    death_elements = soup.find_all('td', class_='death')
    recovered_elements = soup.find_all('td', class_='recovered')
    active_elements = soup.find_all('td', class_='active')

    # Extract the data from the elements and store it in the respective lists
    for i in range(len(state_elements)):
        states.append(state_elements[i].text.strip())
        affected.append(int(affected_elements[i].text.strip()))
        death.append(int(death_elements[i].text.strip()))
        recovered.append(int(recovered_elements[i].text.strip()))
        active.append(int(active_elements[i].text.strip()))

    # Print the source and the number of records collected
    print(f"Data from {source} collected: {len(state_elements)} records")

# Create a bar chart using matplotlib
plt.figure(figsize=(12, 6))
plt.bar(states, active)
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('Active Cases')
plt.title('COVID-19 Active Cases by State in India')
plt.tight_layout()
plt.show()
```

Ministry of Health and Family Welfare, Government of India: https://www.mohfw.gov.in/
